## 用年度、地區分類

```python
import ipywidgets as widgets
from ipywidgets import interact
from ipywidgets import interact_manual
from IPython.display import display

All = 'All'

district = ""
def f(年度,地區):

    if (年度 == 'All') & (地區 == 'All'):
        df_display = df
    elif (地區 == 'All'):
        df_display = df[(df.年度 == 年度)]
    elif (年度 == 'All'):
        df_display = df[(df.洲別 == 地區)]
    else:
        df_display = df[(df.年度 == 年度) & (df.洲別 == 地區)]
    
    df_display.index = range(1,len(df_display)+1) #長度配合欄位數
    display(df_display)
    
    
interact(f, 
         年度=['All',107,106,105],
         地區=['All','歐洲','亞洲','美洲','大洋洲']
        );
```
