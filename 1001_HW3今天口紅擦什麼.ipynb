```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.random import randint
from numpy.random import choice
from IPython.display import display
import ipywidgets as widgets
from ipywidgets import interact
from ipywidgets import interact_manual
```

```python
lip='''
All_fired_up
Dangerous
Flat_out_fabulous
Relentlessly_red
Ruby_woo
Runway_hit
'''
```

```python
phrase1 = lip.split("\n")
l = randint(1,2) 
random = choice(phrase1,l)
today=' '.join(random) 
```

```python
from skimage import io

def lipstick(name):
    if(name == "All_fired_up"):
        image = io.imread("(https://screenshot.net/tw/5dveeho")
    elif(name == "Dangerous"):
        image = io.imread("https://images-na.ssl-images-amazon.com/images/I/51eJ9eqcLNL._SY550_.jpg")
    elif(name == "Flat_out_fabulous"):
        image = io.imread("https://images-na.ssl-images-amazon.com/images/I/51X-JqSCauL._SY550_.jpg")
    elif(name == "Relentlessly_red"):
        image = io.imread("https://images-na.ssl-images-amazon.com/images/I/51eJ9eqcLNL._SY550_.jpg")
    elif(name == "Ruby_woo"):
        image = io.imread("https://cdn11.bigcommerce.com/s-dgbc5lzb0p/images/stencil/1280x1280/products/2828/8271/MAC3351-2__49684.1483656764.jpg?c=2?imbypass=on&imbypass=on")
    elif(name == "Runway_hit"):
        image = io.imread("https://images-na.ssl-images-amazon.com/images/I/516cvXIxa7L._SY355_.jpg")
        
    io.imshow(image)
    io.show()
```

```python
btn_reply= widgets.Button(description='今日推薦')

def btn_eventhandler_reply(obj):
    print(today)
    lipstick(today)

btn_reply.on_click(btn_eventhandler_reply)

display(btn_reply)
```
